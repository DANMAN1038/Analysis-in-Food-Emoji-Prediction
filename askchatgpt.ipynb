```python
# Define a dictionary mapping food-related words to emojis
food_emojis = {
    'grape': '🍇',
    'watermelon': '🍉',
    'banana': '🍌',
    'chili pepper': '🌶️',
    'broccoli': '🥦',
    'bread': '🍞',
    'pancakes': '🥞 ',
    'hamburger': '🍔',
    'french fries': '🍟',
    'pizza': '🍕',
    'hot dog': '🌭',
    'taco': '🌮',
    'popcorn': '🍿',
    'spaghetti': '🍝',
    'cake': '🍰',
    'ice cream': '🍦',
    'doughnut': '🍩',
    'candy': '🍬',
    'cookie': '🍪',
    'pie': '🥧',
}

food_emojis_str = str(food_emojis)

# Define a function that generates a food emoji based on a given text
def generate_food_emoji(text):
    # Extract all the food-related words from the text
    food_words = [word.lower() for word in re.findall(r'\b\w+\b', text) if word.lower() in food_emojis]
    
    # If no food words are found, use the OpenAI API to generate a suggestion for a food-related emoji
    if not food_words:
        prompt = "Suggest three food-related emojis for the following text." + text + "... You must suggest exactly 3 emojis." + "The suggested emojis must be defined in the following dictionary: " + food_emojis_str
        response = openai.Completion.create(
            engine="text-davinci-003",
            prompt=prompt,
            max_tokens=50,
            n=1,
            stop=None,
            temperature=0.5,
        )
        suggestion = response.choices[0].text.strip()
        return suggestion
    
    # Return the food emoji corresponding to the first food word found
    return food_emojis[food_words[0]]
```

```python

f = '/content/tweets refiltered - Predicted Emoji 1.csv'
import csv
import time
import emoji

tot = 0
correct = 0

pred = []
gpt_generated = []

with open(f) as predicted:
  csv_reader = csv.reader(predicted, delimiter=',')
  for row in csv_reader:
    idx = row[0]
    predicted_emoji = row[1]
    tweet = row[2]
    if idx != "0" and idx != "index":
      idx = int(idx)
      if(idx % 60 == 0): # since chatgpt limits us to 60 requests per minute
        time.sleep(60)
      if (predicted_emoji != "N/A"):
        gpt_emoji = generate_food_emoji(tweet)
        pred.append(predicted_emoji)
        gpt_generated.append(gpt_emoji)
        print(idx, 'predicted:', predicted_emoji, 'chatgpt:', gpt_emoji)
        
        tot += 1

          
```

```python


correct = 0
for i in range(len(pred)):
  prediction = emoji.demojize(pred[i])[1:-1]
  gpt = emoji.demojize(gpt_generated[i])
  print(prediction, gpt)
  if prediction in gpt: # since chatgpt might produce a slightly different version of the emoji
    correct += 1

print(correct, 'out of', tot)
```
